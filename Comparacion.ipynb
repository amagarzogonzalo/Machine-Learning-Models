```python
import numpy as np
import matplotlib.pyplot as plt

regresion_lineal = [0.8231079265562025, 0.8324974924774323, 0.7844990548204159, 0.8077858880778589]
xgboost = [0.8378862516793552, 0.8152173913043478, 0.8506616257088847, 0.8325624421831638]
red_neuronal = [0.8481862964621585, 0.8235823582358236, 0.8648393194706995  , 0.8437067773167358]
medidas = ['Acierto', 'Precisin', 'Recall', 'F1']

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Grfico para Acierto
plt.figure(figsize=(8, 6))
plt.bar(['Regresin Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[0], xgboost[0], red_neuronal[0]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Acierto')
plt.show()

# Grfico para Precisin
plt.figure(figsize=(8, 6))
plt.bar(['Regresin Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[1], xgboost[1], red_neuronal[1]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Precisin')
plt.show()

# Grfico para Recall
plt.figure(figsize=(8, 6))
plt.bar(['Regresin Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[2], xgboost[2], red_neuronal[2]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Recall')
plt.show()

# Grfico para F1
plt.figure(figsize=(8, 6))
plt.bar(['Regresin Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[3], xgboost[3], red_neuronal[3]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('F1')
plt.show()

pesos = [0.7, 0.3, 0, 0]  # Acierto, Precisin, Recall, F1

ponderado_regresion_lineal = np.dot(regresion_lineal, pesos)
ponderado_xgboost = np.dot(xgboost, pesos)
ponderado_red_neuronal = np.dot(red_neuronal, pesos)

puntuaciones = [ponderado_regresion_lineal, ponderado_xgboost, ponderado_red_neuronal]

modelos = ['Regresin Lineal', 'XGBoost', 'Red Neuronal']

plt.figure(figsize=(8, 6))
plt.bar(modelos, puntuaciones, color=colors)
plt.xlabel('Modelo')
plt.ylabel('Puntuacin ponderada')
plt.title('Comparacin de Modelos con Mtricas Ponderadas')
plt.show()
```

Recordemos porque es importante la precisin:

Un banco que hace una campaa de marketing con el objetivo de vender depsitos, priorizar la precisin frente a el recall. Esto es porque es ms importante un falso positivo que un falso negativo. El motivo es que una mala precisin es perjudicial para la empresa porque si el banco espera vender ms depsitos de lo que vender realmente, esperar ms beneficios de los que obtendr en realidad (y esto puede manipular y engaar su situacin financiera), mientras que si tiene un recall bajo, aunque la prediccin ha sido mala, afectar de forma positiva a la empresa porque vender ms de lo que se espera.

Si comparamos en un primer lugar con el accuracy, claramente el mejor algoritmo sera la red neuronal. 

Por otro lado, si nos centramos en el accuracy y lo ponderamos con la precisin (medida muy importante para nuestro problema en particular) elegiramos la red neuronal igualmente porque tiene los valores ms altos en accuracy el segundo mayor valor en precisin, solo superado por la regresin. La regresin tiene dos puntos y medio porcentuales menos de accuracy y un punto ms de precisin. Dicho esto, el algoritmo ms conveniente sera la red neuronal ya que al tener un gran acierto y una notable precisin hace de este modelo ideal para el problema. Sin embargo, si se pretendier maximizar al mxmimo la precisin, sin importar perder aciertos, se escogera la regresin logstica renunciando a 2 aciertos y medio menos por cada 100 predicciones a costa de tener un punto ms de precisin.
