```python
import numpy as np
import matplotlib.pyplot as plt

regresion_lineal = [0.8231079265562025, 0.8324974924774323, 0.7844990548204159, 0.8077858880778589]
xgboost = [0.8378862516793552, 0.8152173913043478, 0.8506616257088847, 0.8325624421831638]
red_neuronal = [0.8481862964621585, 0.8235823582358236, 0.8648393194706995  , 0.8437067773167358]
medidas = ['Acierto', 'Precisión', 'Recall', 'F1']

colors = ['#1f77b4', '#ff7f0e', '#2ca02c']

# Gráfico para Acierto
plt.figure(figsize=(8, 6))
plt.bar(['Regresión Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[0], xgboost[0], red_neuronal[0]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Acierto')
plt.show()

# Gráfico para Precisión
plt.figure(figsize=(8, 6))
plt.bar(['Regresión Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[1], xgboost[1], red_neuronal[1]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Precisión')
plt.show()

# Gráfico para Recall
plt.figure(figsize=(8, 6))
plt.bar(['Regresión Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[2], xgboost[2], red_neuronal[2]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('Recall')
plt.show()

# Gráfico para F1
plt.figure(figsize=(8, 6))
plt.bar(['Regresión Lineal', 'XGBoost', 'Red Neuronal'], [regresion_lineal[3], xgboost[3], red_neuronal[3]], color=colors)
plt.xlabel('Modelo')
plt.ylabel('Porcentaje (%)')
plt.title('F1')
plt.show()

pesos = [0.7, 0.3, 0, 0]  # Acierto, Precisión, Recall, F1

ponderado_regresion_lineal = np.dot(regresion_lineal, pesos)
ponderado_xgboost = np.dot(xgboost, pesos)
ponderado_red_neuronal = np.dot(red_neuronal, pesos)

puntuaciones = [ponderado_regresion_lineal, ponderado_xgboost, ponderado_red_neuronal]

modelos = ['Regresión Lineal', 'XGBoost', 'Red Neuronal']

plt.figure(figsize=(8, 6))
plt.bar(modelos, puntuaciones, color=colors)
plt.xlabel('Modelo')
plt.ylabel('Puntuación ponderada')
plt.title('Comparación de Modelos con Métricas Ponderadas')
plt.show()
```

Recordemos porque es importante la precisión:

Un banco que hace una campaña de marketing con el objetivo de vender depósitos, priorizará la precisión frente a el recall. Esto es porque es más importante un falso positivo que un falso negativo. El motivo es que una mala precisión es perjudicial para la empresa porque si el banco espera vender más depósitos de lo que venderá realmente, esperará más beneficios de los que obtendrá en realidad (y esto puede manipular y engañar su situación financiera), mientras que si tiene un recall bajo, aunque la predicción ha sido mala, afectará de forma positiva a la empresa porque venderá más de lo que se espera.

Si comparamos en un primer lugar con el accuracy, claramente el mejor algoritmo sería la red neuronal. 

Por otro lado, si nos centramos en el accuracy y lo ponderamos con la precisión (medida muy importante para nuestro problema en particular) elegiríamos la red neuronal igualmente porque tiene los valores más altos en accuracy el segundo mayor valor en precisión, solo superado por la regresión. La regresión tiene dos puntos y medio porcentuales menos de accuracy y un punto más de precisión. Dicho esto, el algoritmo más conveniente sería la red neuronal ya que al tener un gran acierto y una notable precisión hace de este modelo ideal para el problema. Sin embargo, si se pretendierá maximizar al máxmimo la precisión, sin importar perder aciertos, se escogería la regresión logística renunciando a 2 aciertos y medio menos por cada 100 predicciones a costa de tener un punto más de precisión.
